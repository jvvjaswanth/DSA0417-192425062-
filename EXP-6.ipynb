```python
prices = [100, 200, 50]
quantities = [2, 1, 3]
discount_rate = 10
tax_rate = 5
subtotal = 0
for i in range(len(prices)):
    item_total = prices[i] * quantities[i]
    print("Item", i+1, "Total =", prices[i], "x", quantities[i], "=", item_total)
    subtotal += item_total
print("Subtotal =", subtotal)
discount_amount = (discount_rate / 100) * subtotal
print("Discount (", discount_rate, "% ) =", discount_amount)
after_discount = subtotal - discount_amount
print("Amount After Discount =", after_discount)
tax_amount = (tax_rate / 100) * after_discount
print("Tax (", tax_rate, "% ) =", tax_amount)
total_cost = after_discount + tax_amount
print("Final Total Cost =", total_cost)
```
