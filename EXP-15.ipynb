```python
# Import required libraries
import pandas as pd

# Sample dataset: Number of likes for each post
likes_data = {
    'Post_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Likes': [50, 100, 150, 50, 200, 100, 150, 300, 200, 50]
}

# Create DataFrame
df = pd.DataFrame(likes_data)

# Display the dataset
print("User Interaction Data:")
print(df)

# Calculate frequency distribution of likes
frequency_distribution = df['Likes'].value_counts().sort_index()

# Display frequency distribution
print("\nFrequency Distribution of Likes:")
print(frequency_distribution)
```
