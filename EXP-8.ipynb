```python
import pandas as pd

# Sample DataFrame
data = {
    'product_name': ['Laptop', 'Mouse', 'Laptop', 'Keyboard',
                     'Mouse', 'Laptop', 'Monitor', 'Keyboard'],
    'quantity_sold': [2, 5, 1, 3, 2, 4, 6, 1],
    'order_date': ['2026-01-02', '2026-01-03', '2026-01-05',
                   '2026-01-06', '2026-01-08', '2026-01-10',
                   '2026-01-12', '2026-01-15']
}

sales_data = pd.DataFrame(data)

# Convert order_date to datetime (if needed)
sales_data['order_date'] = pd.to_datetime(sales_data['order_date'])

# ðŸ”¹ Find Top 5 Most Sold Products
top_5_products = (
    sales_data
    .groupby('product_name')['quantity_sold']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

print("Top 5 Most Sold Products:")
print(top_5_products)
```
