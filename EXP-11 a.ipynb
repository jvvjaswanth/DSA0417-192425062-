```python
import matplotlib.pyplot as plt

# Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 150, 180, 200, 170, 220]

# Line Plot
plt.plot(months, sales, marker='o')
plt.title("Monthly Sales Line Plot")
plt.xlabel("Months")
plt.ylabel("Sales (in ₹ thousands)")
plt.grid(True)
plt.show()
```
