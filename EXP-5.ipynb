```python
import numpy as np
fuel_efficiency = np.array([25, 30, 28, 35, 32])
average_efficiency = np.mean(fuel_efficiency)
model_1 = fuel_efficiency[0]
model_4 = fuel_efficiency[3]
percentage_improvement = ((model_4 - model_1) / model_1) * 100
print("Average Fuel Efficiency:", average_efficiency, "mpg")
print("Percentage Improvement from Model 1 to Model 4:", percentage_improvement, "%")
```
