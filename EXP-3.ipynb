```python
import numpy as np
house_data = np.array([
    [3, 1500, 200000],
    [5, 2500, 350000],
    [4, 1800, 250000],
    [6, 3000, 450000],
    [2, 1200, 180000],
    [5, 2700, 400000]
])
filtered_houses = house_data[house_data[:, 0] > 4]
average_price = np.mean(filtered_houses[:, 2])

print("Average Sale Price of Houses with More Than 4 Bedrooms:", average_price)
```
